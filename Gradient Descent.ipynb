Gradient descent for f(x) = 6*x^2 - 12*x + 3

```python
def f(x):
  return 6 * x * x - 12 * x + 3

def d(x):
  return 12 * x - 12

import numpy as np
import matplotlib.pylab as plt

x = np.linspace(-1,3,1000)
plt.plot(x,f(x))

x = -0.9
eta = 0.05 #learning rate
for i in range(15):
  plt.plot(x, f(x), marker ='o', color = 'r')
  x = x - eta * d(x)
```
