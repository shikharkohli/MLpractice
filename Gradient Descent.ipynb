Gradient descent for f(x) = 6*x^2 - 12*x + 3

```python
def f(x):
  return 6 * x * x - 12 * x + 3

def d(x):
  return 12 * x - 12

import numpy as np
import matplotlib.pylab as plt

x = np.linspace(-1,3,1000)
plt.plot(x,f(x))

x = -0.9
eta = 0.05 #learning rate
for i in range(15):
  plt.plot(x, f(x), marker ='o', color = 'r')
  x = x - eta * d(x)
```

```python

import numpy as np
import matplotlib.pylab as plt

# two dimensional gradient descent

def f(x,y):
  return 6 * x * x + 9 * y * y - 12 * x - 14 * y +3

def dx(x,y):
  return 12 * x - 12

def dy(x,y):
  return 18 * y - 14

N = 100

x,y = np.meshgrid(np.linspace(-1, 3, N), np.linspace(-1, 3, N))

z = f(x,y)

plt.contour(x,y,z, 10, cmap = "Greys")
plt.contour(x,y,z,10, colors = 'k', linewidths=1)
plt.plot([0,0],[-1,3], color = 'k', linewidth = 1)
plt.plot([-1,3],[0,0], color = 'k', linewidth = 1)

plt.plot(1, 0.7777778, color = 'k', marker = '+')
x = ptx = -1
y = pty = -1

x_values = []
y_values = []

for i in range (0,N):
  x_values.append(x)
  y_values.append(y)
  plt.plot([ptx, x], [pty, y], marker='o', linestyle='dotted', color='k')
  ptx = x
  pty = y
  x = x - eta * dx(x,y)
  y = y - eta * dy(x,pty)
  plt.plot(ptx, pty, color = 'r', marker = 'o')

#Plot the changing x and y values
plt.plot(x_values, y_values, marker='x', linestyle='-', color='b', label='Gradient Descent Path')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Contour Plot with Gradient Descent Path')
plt.legend()
plt.show()
```
